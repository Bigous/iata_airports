```python
import numpy as np
import pandas as pd
```

```python
# url = 'https://github.com/jpatokal/openflights/raw/master/data/airports.dat'
url = 'https://davidmegginson.github.io/ourairports-data/airports.csv'

df = pd.read_csv(url)

df
```

```python
airports = df[df["Type"] == "airport"]
airports
```

```python
from timezonefinder import TimezoneFinder
tf = TimezoneFinder()
```

```python
tf.timezone_at(lat=-23.55,lng=-46.63)
```

```python
tf.timezone_at(lat=-6.081690, lng=145.391998)
```

```python
airports[airports["Tz database time zone"] == None]
```

```python
def replace_invalid_tz(row):
    tz = row['Tz database time zone']
    if pd.isna(tz) or tz == "\\N":
        return tf.timezone_at(lng=row['Longitude'], lat=row['Latitude'])
    else:
        tz

airports['Tz database time zone'] = df.apply(replace_invalid_tz, axis=1)
airports
```
