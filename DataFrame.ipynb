```python
import numpy as np
import pandas as pd
```

```python
# url = 'https://github.com/jpatokal/openflights/raw/master/data/airports.dat'
url = 'https://davidmegginson.github.io/ourairports-data/airports.csv'

df = pd.read_csv(url)

df
```

```python
df['type'].unique()
```

```python
iata_airports = df[df["type"].isin(['small_airport', 'medium_airport', 'large_airport']) & df['iata_code'].notna()].copy()
iata_airports
```

```python
from timezonefinder import TimezoneFinder
tf = TimezoneFinder()
```

```python
tf.timezone_at(lat=-23.55,lng=-46.63)
```

```python
iata_airports['timezone'] = iata_airports.apply(lambda row: tf.timezone_at(lat=row['latitude_deg'], lng=row['longitude_deg']), axis=1)
iata_airports
```

```python
iata_airports.to_csv('iata_airports.csv', index=False)
```
